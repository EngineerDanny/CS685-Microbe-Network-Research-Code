```python
# Import pandas and plotnine
import pandas as pd
from plotnine import *

# Create a pandas DataFrame from the data
df = pd.read_csv("/projects/genomic-ml/da2343/ml_project_1/big_o/results.csv")

# Plot the mean time and mean memory for each function and N value using plotnine
p = (ggplot(df, aes(x="N", y="Mean Time", color="Function"))
     + geom_point()
     + geom_smooth(method="lm")
     + scale_x_log10()
     + scale_y_log10()
     + labs(x="N", 
            y="Mean Time (s)", 
            title="Time Performance Comparison of Different Functions")
     + theme_minimal()
    )

# Show the plot
p
```
