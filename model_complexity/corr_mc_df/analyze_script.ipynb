```python
import pandas as pd

df = pd.read_csv('/projects/genomic-ml/da2343/ml_project_1/model_complexity/corr_mc_df/2023-10-23_11:59.csv')
df_pearson = df[(df['algorithm'] == 'Pearson') & (df['fold_id'] == 2 ) ]
# df_pearson = df[(df['algorithm'] == 'Pearson')]
# df_spearman = df[df['algorithm'] == 'Spearman' & df['fold_id'] == 2]

df = df_pearson
df
```

```python
# Performing the calculations
result_df = df.groupby('reg_param').agg({
    'edges': 'mean',
    'subtrain_score': ['mean', 'std'],
    'validation_score': ['mean', 'std']
}).reset_index()

result_df.columns = result_df.columns.map('_'.join)
```
